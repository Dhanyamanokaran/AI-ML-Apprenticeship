```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 

data={
    "Name":["Arun", "Priya", "Karthik", "Divya", "Vignesh", "Anitha"],
    "StudyHours":[4,6,8,7,7,3],
    "Marks":[45,78,88,56,90,35]
             }
df = pd.DataFrame(data)
print(df)
```

```python
plt.plot(df["StudyHours"],df["Marks"],marker="o",label = "Study Vs Marks")
plt.title("Study Hours vs Marks")
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.legend()
plt.show()
```

```python
plt.bar(df["StudyHours"],df["Marks"],label = "Study Vs Marks",color = "Skyblue")
plt.title("Study Hours vs Marks")
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.legend()
plt.show()
```

```python
plt.hist(df["Marks"],bins = 5,color = "Orange",edgecolor="black")
plt.title("Marks Distribution")
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.barplot(x= "Name",y = "Marks",data=df)
plt.title("Student Marks")
plt.show()
```

```python
sns.histplot(df["Marks"], bins=5, kde=True)
plt.title("Seaborn: Marks Distribution")
plt.show()
```

```python
sns.scatterplot(x="StudyHours", y="Marks", data=df, hue="Name", s=100)
plt.title("Study Hours vs Marks (Seaborn)")
plt.show()
```
